```python
import glob
import pandas as pd
import numpy as np
```

```python
# Initialise a dataframe to store data
bike_df = pd.DataFrame()

# List of tripdata csv file pathnames, that are concatenated into bike dataframe
data_list = [str(i) for i in glob.glob('data\JC-2016??-citibike-tripdata.csv')]
bike_df = pd.concat([pd.read_csv(file) for file in data_list])

# Initial cleaning and reordering of columns
bike_df.reset_index(drop=True, inplace=True)
bike_df['ID'] = bike_df.index 
bike_df = bike_df[['ID', 'Trip Duration', 'Start Time', 'Start Station ID',
       'Start Station Name', 'Start Station Latitude',
       'Start Station Longitude', 'Stop Time', 'End Station ID', 'End Station Name',
       'End Station Latitude', 'End Station Longitude', 'Bike ID', 'User Type',
       'Birth Year', 'Gender']]
```

- Start and Stop times should be converted to DATETIME types.

```python
bike_df['Start Time'] = pd.to_datetime(bike_df['Start Time'])
bike_df['Stop Time'] = pd.to_datetime(bike_df['Stop Time'])
bike_df.dtypes
```

```python
# Remove spaces from column names
bike_df.columns = [x.replace(' ', '_').lower() for x in bike_df.columns]
print(bike_df.columns)
```

### Found Issues to investigate
- User type has 380 missing values.
- Birth Year has a unexpectadly small minimum value (1900) and 19 thousand missing values
- Trip duration has an extremely high max value (5 orders of magnitude greater than it's 75th percentile value)
- Gender does not appear to be missing data but values of 0 indicated an unknown gender, these might need to be removed for gender based analysis/ dealt with.

### Missing Value & Outlier Analysis


```python
# Find list of lowest birth years
bike_df['age'] = 2016 - bike_df.birth_year.values
bike_df[['id', 'birth_year', 'age']].sort_values('age', ascending=False)[:10]

```

- There seems to be a single outlier birth year of 1900 (age = 116) under index 215055 that can be dropped
- It will be unexpected to encounter further riders of that age, so the observation can be safely dropped

```python
# Drop observation in place and inspect the new top ten highest age rider records
bike_df.drop([215055], inplace=True)
bike_df[['id', 'birth_year', 'age']].sort_values('age', ascending=False)[:10]
```

```python
# First 10 missing birth year records
missing_birth_year_observations = bike_df[bike_df.birth_year.isnull()]
missing_birth_year_observations.head(10)
```

#### Observations
- Gender seems to always be 0 (unknown)
- User type seems to always be 'Customer'


```python
print("Missing birth year obvservations by Customer vs Subscriber 'user_type': ")
print(missing_birth_year_observations.user_type.value_counts(dropna=False))
print('\n')


print("Percentage of each user type with missing birth year observation:")
print(missing_birth_year_observations.user_type.value_counts() / bike_df.user_type.value_counts())
print('\n')

print('Number of missing birth years per gender type:')
print(missing_birth_year_observations.gender.value_counts())
```

- The majority of missing birth year observations are from Customers ('Customer' user_type)
    - 99.67% of Customers are missing birth year data
    - 1.52% of Subscribers are missing birth year data  
    - Therefore it is inadvisable to remove all observations with missing birth year data, as this would remove all customers. This will have to be worked around when developing database schema/ views
    
<br>


- All missing birth year observations have 0 recorded for gender (unknown gender)

```python
missing_user_type_observation = bike_df[bike_df.user_type.isnull()]

print('Number of no user_type observations that also have no birth year:')
print(missing_user_type_observation.birth_year.isna().sum())
print('\n')

print('Number of no user_type observations that also have an unknown (0) gender:')
print(len(missing_user_type_observation[missing_user_type_observation.gender == 0]))

missing_user_type_observation.head()
```

- Observations with missing user_type do not have missing birth years or unknown genders
- Therefore it is likely that these records are subscriber records and not customer records, as we previously found that customer records are 99.7% likely to have a missing birth year.

```python
my = (missing_user_type_observation.resample('M', on='start_time').count()['id'] / len(missing_user_type_observation)).rename('missing')
dy = (bike_df.resample('M', on='start_time').count()['id'] / len(bike_df)).rename('full')
combo = pd.concat([my,dy],axis=1)
ax = combo.plot.bar();
```

- The missing user_type data is a small percentage of total observations (~0.15%) however based on the time distribution of the data missingness, it does not seem to be missing at random
- Thus it is preferable to replace the NaN values with an 'unknown' category, which can be analysed/ investigated further

```python
bike_df.fillna({'user_type' : 'Unknown'}, inplace=True)
bike_df.isna().sum()
```

```python
bike_df.groupby(['user_type', 'gender']).count()['id']
```

- The majority (~75%) of observations with unknown (0) gender, belong to the 'Customer' user_type. These rows cannot be deleted (they comprise almost every single customer observation), but must be taken into account when attempting to use gender information 
- I.e. gender will only be useful for analysing subscriber information, not customers.

### Trip duration Analysis

- There was a suspisiously large max value in trip duration (5 orders of magnitude larger than the 75th percentile value)

```python
# Create two columns in the dataframe to indicate trip minutes and hours
bike_df['trip_minutes'] = round(bike_df['trip_duration'] / 60, 2)
bike_df['trip_hours'] = round(bike_df['trip_duration'] / (60**2), 2)
bike_df[['trip_duration','trip_minutes','trip_hours']].sort_values('trip_duration', ascending=False)[:20]
```

```python
bike_df[['trip_duration','trip_minutes','trip_hours']].describe()
```

- Citybike policy indicates that the maximum length of a trip is 24 hours, however it seems there is a very large number of trips that violate this policy, these observations will be left within the dataset to allow for analysis of how these rules were broken.
- Instead we will add a flag to highlight observations that violate these rules, such that they can be filtered if neccessary.

```python
bike_df['valid_duration'] = bike_df['trip_duration'].apply(lambda x: 0 if x > 86400 else 1).astype(bool) #24 hours is 86400 seconds
```

### Final Cleanup

```python
bike_df.reset_index(drop=True, inplace=True)
bike_df.id = bike_df.index
bike_df = bike_df[['id', 'start_time', 'stop_time', 'trip_duration', 'trip_minutes',
       'trip_hours','start_station_id',
       'start_station_name', 'start_station_latitude',
       'start_station_longitude', 'end_station_id',
       'end_station_name', 'end_station_latitude', 'end_station_longitude',
       'bike_id', 'user_type', 'birth_year', 'gender', 'valid_duration']]
bike_df.head()
```
